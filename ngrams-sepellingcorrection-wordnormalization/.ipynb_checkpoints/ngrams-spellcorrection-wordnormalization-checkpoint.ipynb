```python
def build_api():
    """Build api untuk autentikasi agar terhubung ke twitter"""
    consumer_key = ""
    consumer_secret = ""
    access_token = ""
    access_token_secret = ""
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    return tweepy.API(auth)
```

```python
def remove_empty_line(sentences):
    """Untuk hapus baris yang berisi character enter"""
    sentences = sentences.split("\n")
    non_empty_line = [line for line in sentences if line != ""]
    result = ""
    for line in non_empty_line:
        result += line + "\n"
    return result
```

```python
def remove_character(sentences):
    """untuk menghapus link, angka dan character yang tidak diperlukan."""
    rgx_url = 'https://t.co/[\w]*[\W]*'
    rgx_user = '[!@()0-9`,-/"?_‘’#]+|&amp;+|:|^ | $'

    sentences = re.sub(rgx_url, '', sentences)
    sentences = re.sub(rgx_user, '', sentences)
    return sentences
```

```python
def ordered_tokenize(sentences):
    """Untuk melakukan tokenisasi. supaya tidak terdapat kata yang berulang"""
    tokens = nltk.word_tokenize(sentences)
    result = []
    ordered_tokens = set()
    for word in tokens:
        if word not in ordered_tokens:
            ordered_tokens.add(word)
            result.append(word)
    return result
```

```python
def words(text):
    return re.findall(r'\w+', text.lower())
```

```python
WORDS = Counter(words(open('kata-dasar.txt').read()))
def P(word, N=sum(WORDS.values())):
    # "Probability of `word`."
    return WORDS[word] / N
```

```python
def correction(word):
    # "Most probable spelling correction for word."
    return max(candidates(word), key=P)
```

```python
def candidates(word):
    # "Generate possible spelling corrections for word."
    return (known([word])) or known(edits1(word) or known(edits2(word)) or [word])
```

```python
def known(words):
    # "The subset of `words` that appear in the dictionary of WORDS."
    return set(w for w in words if w in WORDS)
```

```python
def edits1(word):
    # "All edits that are one edit away from `word`."
    letters = 'abcdefghijklmnopqrstuvwxyz'
    splits = [(word[:i], word[i:]) for i in
              range(len(word) + 1)]  # [('', 'kemarin'), ('k', 'emarin'), ('ke', 'marin'), dst]
    deletes = [L + R[1:] for L, R in splits if R]  # ['emarin', 'kmarin', 'kearin', dst]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R) > 1]  # ['ekmarin', 'kmearin', 'keamrin', dst]
    replaces = [L + c + R[1:] for L, R in splits if R for c in letters]  # ['aemarin', 'bemarin', 'cemarin', dst]
    inserts = [L + c + R for L, R in splits for c in letters]  # ['akemarin', 'bkemarin', 'ckemarin', dst]
    return set(deletes + transposes + replaces + inserts)
```

```python
def edits2(word):
    # "All edits that are two edits away from `word`."
    return (e2 for e1 in edits1(word) for e2 in edits1(e1))
```

```python
stemmer = StemmerFactory().create_stemmer()

# baca file yang telah diisi oleh data streaming twitter
# lakukan penghapusan baris yang berisi character enter dan character-character yang tidak dibutuhkan
sentences_rewrite = remove_character(remove_empty_line(open("tweet_data.txt", "r").read().lower()))

# file yang telah dilakukan pengeditan sebelumnya di kembalikan ke bentuk kata dasar nya.
# lalu tulis ulang pada file tweet_data_rewrite.txt
sentences = open('tweet_data_rewrite.txt', 'w')
kata_baku = stemmer.stem(sentences_rewrite)
tokens = ordered_tokenize(kata_baku)
for token in tokens:
    sentences.write(token+" ")

sentences.close()
print("success to rewrite data")
```

```python
# Create 7 grams
sentences = open("tweet_data_rewrite.txt", "r").read()
sevenGrams = ngrams(sentences.split(), 7)
for grams in sevenGrams:
    print(grams)
```

```python
print('{:^24s}'.format("kata typo"), '{:^24s}'.format("koreksi"), '{:^24s}'.format("Hasil"))
print("------------------------------------------------------------------------------------")
for kata in sentences.split():
    # print(kata+"\n")
    try:
        hasil = "Gagal"
        if kata.__eq__(correction(kata)): hasil = "Benar"

        print('{:^24s}'.format(kata), end="")
        print('{:^24s}'.format(correction(kata)), end="")
        print('{:^24s}'.format(hasil), end="\n")
    except:
        print('{:^24s}'.format(kata), end="")
        print('{:^24s}'.format("-"), end="")
        print('{:^24s}'.format("Error proses"), end="\n")
```
